```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
plt.bar(df["age"], df["annual_income"])
plt.xlabel("Age")
plt.ylabel("Annual Income")
plt.title("Age vs Annual Income")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
sns.boxplot(x=df["region"], y=df["purchase_frequency"])
plt.xlabel("Region")
plt.ylabel("Purchase Frequency")
plt.title("Region vs Purchase Frequency")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
sns.lineplot(x=df["purchase_amount"], y=df["loyalty_score"])
plt.xlabel("Purchase amount")
plt.ylabel("Loyalty score")
plt.title("Purchase amount vs Loyalty_score")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
plt.bar(df["age"], df["annual_income"], color='red')
plt.xlabel("Age")
plt.ylabel("Annual Income")
plt.title("Age vs Annual Income")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
plt.barh(df["age"], df["annual_income"], color='red')
plt.xlabel("Age")
plt.ylabel("Annual Income")
plt.title("Age vs Annual Income")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
sns.histplot(x=df["purchase_amount"], kde=True)
plt.xlabel("Purchase amount")
plt.ylabel("Frequency")
plt.title("Purchase amount distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
plt.fill_between(df["purchase_amount"], df["loyalty_score"], color='skyblue', alpha=0.4)
plt.plot(df["purchase_amount"], df["loyalty_score"], color='Slateblue', alpha=0.6)
plt.xlabel("Purchase Amount")
plt.ylabel("Loyalty Score")
plt.title("Purchase Amount vs Loyalty Score (Area Chart)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
sns.scatterplot(x=df["purchase_amount"], y=df["loyalty_score"], color='blue')
plt.xlabel("Purchase Amount")
plt.ylabel("Loyalty Score")
plt.title("Purchase Amount vs Loyalty Score (Scatter Plot)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
region_counts = df.groupby("region")["purchase_frequency"].sum()
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Purchase Frequency Distribution by Region")
plt.axis('equal')
plt.show()
```
