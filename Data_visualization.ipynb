```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("sample_data/Customer Purchasing Behaviors.csv")
region_counts = df.groupby("region")["purchase_frequency"].sum()
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Purchase Frequency Distribution by Region")
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
temprature=np.array([22,25,28,30,27,26,24,23,25,22,21,20])
mean_temprature=np.mean(temprature)
print("Mean:",mean_temprature)
median_temprature=np.median(temprature)
print("Median:",median_temprature)
mode_temprature=stats.mode(temprature)
print("Mode:",mode_temprature)
std_temprature=np.std(temprature)
print("Standard Deviation:",std_temprature)
var_temprature=np.var(temprature)
print("Variance:",var_temprature)
plt.hist(temprature)
plt.show()
plt.hist(mode_temprature)
plt.show()
```
